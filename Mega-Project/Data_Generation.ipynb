```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
from google.colab import drive
drive.mount('/content/gdrive/', force_remount=True)
```

```python
from itertools import permutations
comb = permutations([1,2,3,4],4)
mat=[]
for num in comb:
    mat.append(list(num))
```

```python
for i in range(21,42):
  img = np.array(Image.open('/content/gdrive/MyDrive/train_images/{0:0=2d}_training.tif'.format(i)))

  first = np.array(img[170:190, 115:160, :])
  second = np.array(img[115:135, 340:385, :])
  third = np.array(img[355:375, 115:160, :])
  fourth = np.array(img[435:455, 340:385, :])
  dic={1:first,2:second,3:third,4:fourth}

  fnm=0
  for com in mat:
    fnm+=1
    src=img[:,:,:]
    src[170:190, 115:160, :] = dic[com[0]]
    src[115:135, 340:385, :] = dic[com[1]]
    src[355:375, 115:160, :] = dic[com[2]]
    src[435:455, 340:385, :] = dic[com[3]]
    im = Image.fromarray(src)
    im.save("/content/Data/{0:0=2d}_{1:0=2d}_traning.png".format(fnm,i))
```
