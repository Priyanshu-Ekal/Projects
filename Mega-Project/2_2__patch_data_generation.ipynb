```python
from itertools import permutations
```

```python
comb = permutations([1,2,3,4],4)
```

```python
mat=[]
for num in comb:
    mat.append(list(num))
```

```python
dic={1:first,2:second,3:third,4:fourth}
```

```python
fnm=0
for com in mat:
  fnm+=1
  src=img[:,:,:]
  src[166:295, 52:282, :] = dic[com[0]]
  src[166:295, 282:512, :] = dic[com[1]]
  src[295:424, 52:282, :] = dic[com[2]]
  src[295:424, 282:512, :] = dic[com[3]]

  im = Image.fromarray(src)
  im.save("/content/Data/40_traning_{0:0=2d}.tif".format(fnm))
```
