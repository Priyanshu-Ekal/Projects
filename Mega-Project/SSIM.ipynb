```python
def display_images(image1, image2):
    fig, axes = plt.subplots(1, 2, figsize=(5, 5))
    
    axes[0].imshow(image1)
    axes[0].axis('off')
    
    axes[1].imshow(image2)
    axes[1].axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
def calculate_ssim(real_images,generated_images):
          score, _ = ssim(real_images, generated_images, win_size=3, full=True)
          print("SSIM SCORE OF REAL AND GENERATED IMAGE: %.2f" % (score))
          
```
